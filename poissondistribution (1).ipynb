```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

lam = 38
k = 25

# probability
prob = poisson.pmf(k, lam)

print("Probability of getting 15 calls in a minute:", prob)

# distribution
x = np.arange(0, 21)
fit = poisson.pmf(x, lam)

plt.bar(x, fit, color="lightgreen", edgecolor="black")
plt.title("Poisson Distribution")
plt.xlabel("Number of calls per minute")
plt.ylabel("Probability")

plt.show()
```
