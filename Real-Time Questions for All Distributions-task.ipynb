```python
    BINOMIAL DISTRIBUTION:

from scipy.stats import binom

# Parameters
n = 10       # total items
p = 0.95     # probability of good item
k = 9        # exactly 9 good items

# Probability
probability = binom.pmf(k, n, p)

print(probability)
```

```python
    CONFIDENCE INTERVAL:

import math

mean = 165
sd = 10
n = 50
z = 1.96

se = sd / math.sqrt(n)
margin = z * se

lower = mean - margin
upper = mean + margin

print(lower, upper)
```

```python
import math

mean = 500
sd = 8
n = 40
z = 2.576   # 99% confidence

se = sd / math.sqrt(n)
margin = z * se

lower = mean - margin
upper = mean + margin

print("99% Confidence Interval:", (lower, upper))
```
