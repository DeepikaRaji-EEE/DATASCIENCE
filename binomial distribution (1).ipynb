```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Parameters
n = 12
p = 0.7
k = 10

# Exact probability
prob = binom.pmf(k, n, p)

# Generate PMF values
x = np.arange(0, n + 1)
pmf_values = binom.pmf(x, n, p)

# Plot full PMF
plt.bar(x, pmf_values,
        color="yellow",
        edgecolor="black",
        label="Binomial PMF")

# Highlight k = 10
plt.bar(k, prob,
        color="green",
        edgecolor="black",
        label=f"k = {k}")

plt.xlabel("Number of Successes")
plt.ylabel("Probability")
plt.title("Binomial Distribution")

plt.legend()
plt.grid(axis="y", alpha=0.3)

plt.show()
```

```python
from scipy.stats import binom

n = 12
p = 0.9
k = 10

cdf_value = binom.cdf(k, n, p)

print("P(X ≤ 10):", cdf_value)
```
