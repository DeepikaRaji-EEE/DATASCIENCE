```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Simulated real-time data
stream_data = np.random.normal(100, 10, 200)

# Add anomaly
stream_data = np.append(stream_data, 300)

# Normal Q-Q Plot
stats.probplot(stream_data, dist="norm", plot=plt)

plt.title("Normal Q-Q Plot - Real Time Data")
plt.show()
```

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Simulated real-time exponential data
data = np.random.exponential(scale=1, size=200)

# Exponential Q-Q Plot
stats.probplot(data, dist=stats.expon, plot=plt)

plt.title("Exponential Q-Q Plot")
plt.show()
```

```python

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Simulated real-time uniform data
data = np.random.uniform(0, 1, 200)

# Uniform Q-Q Plot
stats.probplot(data, dist=stats.uniform, plot=plt)

plt.title("Uniform Q-Q Plot")
plt.show()
```
