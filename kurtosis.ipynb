```python
import numpy as np
from scipy.stats import kurtosis as k
profits=([10000,1200,1500,200,40,100,200,55,39,2,6,5,4,3,53,1,2])
kurto=k(profits,fisher=True)
kurto1=k(profits,fisher=False)
mean_value=np.mean(profits)
max_value=max(profits)
min_value=min(profits)
if(kurto>0):
    print("heavy tail more distribution")
    if(max_value-mean_value)>(mean_value-min_value):
        print("big profits")
    else:
        print("big loss")
elif(kurto<0):
    print("low tail less distribution and low profit/loss")
else:
    print("normal distribution and normal profit/loss")
```

```python
import matplotlib.pyplot as plt
plt.hist(profits, color="red", edgecolor="black")
plt.title("Profit or Loss Distribution")
plt.xlabel("Profit Values")
plt.xlabel("loss Values")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.plot(profits, color="red")
plt.title("Profit or Loss Distribution")
plt.xlabel("Profit Values")
plt.ylabel("Frequency")
plt.axhline(mean_val, linestyle="--", linewidth=3)
plt.axhline(median_val, linestyle="dotted", linewidth=2)
plt.show()
```

```python
import numpy as np
from scipy.stats import kurtosis as k

# Platykurtic data (values spread out evenly)
sales = [1500, 1200, 1400, 1800, 2000, 2390, 2580, 2754, 3023]

kurto = k(sales, fisher=True)

print("Kurtosis Value =", kurto)

if kurto < 0:
    print("Platykurtic → Light tail, low variation")
else:
    print("Not Platykurtic")
```

```python
import numpy as np
from scipy.stats import kurtosis as k

# Mesokurtic data (normal distribution-like)
heights = [1800, 2000, 2390, 16, 15, 14, 12, 10]
kurto = k(heights, fisher=True)

print("Kurtosis Value =", kurto)

# Since perfect 0 is rare, we check "close to zero"
if kurto < 0.1:
    print("Mesokurtic → Normal distribution")
else:
    print("Not Mesokurtic")
```
