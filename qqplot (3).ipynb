```python
import numpy as np
import matplotlib.pyplot as plt

# Generate normal distribution data
data = np.random.normal(0, 2, 1000)

# Plot histogram
plt.hist(data, bins=20, color='skyblue', edgecolor='black')

plt.title("Normal Distribution")
plt.xlabel("Values")
plt.ylabel("Frequency")

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate uniform distribution data
data = np.random.uniform(0, 4, 1000)

# Plot histogram
plt.hist(data, bins=40, color='orange', edgecolor='black')

plt.title("Uniform Distribution")
plt.xlabel("Values")
plt.ylabel("Frequency")

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate exponential distribution data
data = np.random.exponential(10, 1000)

# Plot histogram
plt.hist(data, bins=20, color='lightgreen', edgecolor='black')

plt.title("Exponential Distribution")
plt.xlabel("Values")
plt.ylabel("Frequency")

plt.show()
```
