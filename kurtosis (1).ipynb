```python
import numpy as np
import matplotlib.pyplot as plt

# sample profit/loss data
profits = np.random.normal(0, 50, 100)

mean_val = np.mean(profits)
median_val = np.median(profits)

plt.figure(figsize=(8,5))

# line plot (sequence view)
plt.plot(profits, color="red", label="Profit/Loss over time")

# reference lines
plt.axhline(mean_val, linestyle="--", linewidth=3, color="blue", label="Mean")
plt.axhline(median_val, linestyle="dotted", linewidth=2, color="green", label="Median")

plt.title("Profit or Loss Trend")
plt.xlabel("Time / Index")
plt.ylabel("Profit Values")

plt.legend()
plt.show()
```

```python
import numpy as np
from scipy.stats import kurtosis as k

# Platykurtic data (values spread out evenly)
sales = [1500, 1200, 1400, 1800, 2000, 2390, 2580, 2754, 3023]

kurto = k(sales, fisher=True)

print("Kurtosis Value =", kurto)

if kurto < 0:
    print("Platykurtic → Light tail, low variation")
else:
    print("Not Platykurtic")
```

```python
import numpy as np
from scipy.stats import kurtosis as k

# Mesokurtic data (normal distribution-like)
heights = [1800, 2000, 2390, 16, 15, 14, 12, 10]
kurto = k(heights, fisher=True)

print("Kurtosis Value =", kurto)

# Since perfect 0 is rare, we check "close to zero"
if kurto < 0.1:
    print("Mesokurtic → Normal distribution")
else:
    print("Not Mesokurtic")
```
