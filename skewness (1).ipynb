```python
import numpy as np
import matplotlib.pyplot as plt

# sample data (example: student roll distribution)
data = np.random.randint(1, 100, 200)

plt.figure(figsize=(8,5))
plt.hist(data, color='lightblue', bins=10, edgecolor='black')

plt.title("Roll Distribution")
plt.xlabel("Roll")
plt.ylabel("Frequency")
plt.grid(True)

plt.show()
```
